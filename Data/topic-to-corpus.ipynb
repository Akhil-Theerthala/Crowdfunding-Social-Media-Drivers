```python
import pandas as pd

final_data = pd.read_csv('Kickstarter_only/processed_data_with_indices.csv')
post_data = pd.read_csv('Kickstarter_only/2022-07-22-09-31-10-IST-feed-download.csv')
```

```python
post_data.keys()
```

```python
from tqdm import tqdm
# final_data=final_data.set_index('index')

corpus_with_topics = final_data.index.values.tolist()
topic_cols = ['topic_0', 'topic_1', 'topic_2', 'topic_3', 'topic_4', 'topic_5', 'topic_6', 'topic_7', 'topic_8', 'topic_9', 'topic_10']

content = []

for i in tqdm(corpus_with_topics):
    curr_message = post_data.loc[i, 'Message']
    curr_link_text = post_data.loc[i, 'Link Text']
    curr_desc = post_data.loc[i, 'Description']
    curr_topics = final_data.loc[i, topic_cols]
    content.append([curr_message, curr_link_text, curr_desc, *curr_topics])

```

```python
pd.DataFrame(content, columns=['Message', 'Link Text', 'Description', *topic_cols]).to_csv('Kickstarter_only/processed_data_with_topics.csv', index=False)
```
